```python
concepts = wc.search( query = "2022年8月后上市 概念 主营").fillna('')

concepts_list = pd.read_csv('./行业概念/Concepts_List.csv',encoding="gbk")
industry_concepts_list = concepts_list.loc[concepts_list['分类'].isin(['行业'])]['概念名称'].values.tolist()
events_concepts_list = concepts_list.loc[concepts_list['分类'].isin(['事件','知名企业','农村','改革','军工','国企体系'])]['概念名称'].values.tolist()
trading_concepts_list = concepts_list.loc[concepts_list['分类'].isin(['交易'])]['概念名称'].values.tolist() # 区域missing


concepts['行业概念'] = ''
concepts['事件概念'] = ''
concepts['交易概念'] = ''

concept_list = []

for row, l in concepts.iterrows():
    concepts['行业概念'][row] = ','.join(set(l['所属概念'].replace('概念','').split(';')).intersection(industry_concepts_list))
    concepts['事件概念'][row] = ','.join(set(l['所属概念'].replace('概念','').split(';')).intersection(events_concepts_list))
    concepts['交易概念'][row] = ','.join(set(l['所属概念'].replace('概念','').split(';')).intersection(trading_concepts_list))
    concept_list = concept_list + l['所属概念'].replace('概念','').split(';')

concepts['a股市值(不含限售股)'] = pd.to_numeric(concepts['a股市值(不含限售股)'])/100000000

code = concepts.pop('股票代码')
name = concepts.pop('股票简称')

concepts = concepts.drop(columns=['公司网站','经营范围','所属概念','主营产品名称','所属概念数量']).round(2).\
            rename(columns = {'a股市值(不含限售股)':'流通市值'})
concepts.insert(0,'股票简称',name)
concepts.insert(0,'股票代码',code)
```

```python
# 当日涨停数据
_percent = wc.search( query = "涨停").fillna('')
#_percent.to_csv('zhangting.csv',index = False,encoding="gbk")
```

```python
# 仅做分析 不合并入最终表格
_percent['流通市值'] = pd.to_numeric(_percent['a股市值(不含限售股)'])/100000000
_percent['封单额'] = pd.to_numeric(_percent['涨停封单额'])/10000000
_percent['连续涨停天数'] = pd.to_numeric(_percent['连续涨停天数'])
_percent['封单成交比'] = pd.to_numeric(_percent['涨停封单量占成交量比'])

_percent.sort_values(['涨停封单量占成交量比','连续涨停天数','流通市值'],ascending = False, inplace=True)
_percent[['股票代码','股票简称','最新涨跌幅','流通市值','涨停类型','连续涨停天数','几天几板','封单额','封单成交比','首次涨停时间','最终涨停时间','涨停开板次数']] \
            .round(2).loc[_percent['股票简称'].str.contains('ST') == False]
```

```python
# 所属指数数据
China_index = wc.search( query = "2022年8月后上市 所属指数类 市值 几天几板").fillna('')

#index_selected_list = ['上证50','上证150','上证380','中证100','中证200','中证500','中证800',\
#                       '沪深300','科创50','中证1000','创业300']

index_list = pd.read_csv('./行业概念/index_list.csv',encoding="gbk")
industry_index_list = index_list.loc[index_list['分类'].isin(['行业'])]['指数名称'].values.tolist()
hushen_index_list = index_list.loc[index_list['分类'].isin(['中证','沪深','上证','深证'])]['指数名称'].values.tolist()
style_index_list = index_list.loc[index_list['分类'].isin(['风格'])]['指数名称'].values.tolist()
# all_index = []

China_index['沪深指数'] = ''
China_index['行业指数'] = ''
China_index['风格指数'] = ''

for row, l in China_index.iterrows():
    China_index['沪深指数'][row] = ','.join(set(l['所属指数类'].split(';')).intersection(hushen_index_list))
    China_index['行业指数'][row] = ','.join(set(l['所属指数类'].split(';')).intersection(industry_index_list))
    China_index['风格指数'][row] = ','.join(set(l['所属指数类'].split(';')).intersection(style_index_list))
#    all_index = all_index + l['所属指数类'].split(';')
```

```python

China_index['最新涨跌幅'] = pd.to_numeric(China_index['最新涨跌幅'])
China_index['流通市值'] = pd.to_numeric(China_index['a股市值(不含限售股)'])/100000000
China_index.sort_values(['最新涨跌幅'], ascending = False).round(2) \
                       [['股票代码','股票简称','最新价','最新涨跌幅','流通市值','几天几板','沪深指数','风格指数','行业指数']]#.to_csv('index.csv',index = False ,encoding="gbk")
                      
```
