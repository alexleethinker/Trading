```python
import requests
import pandas as pd
import json
from tqdm import tqdm

def get_operateDepart():
    url = 'https://datacenter-web.eastmoney.com/api/data/v1/get'
    df = pd.DataFrame()
    for i in tqdm(range(1,10)):
        params = {
            'sortColumns': 'AMOUNT,OPERATEDEPT_CODE',
            'sortTypes': '-1,1',
            'pageSize': '500',
            'pageNumber': str(i),
            'reportName': 'RPT_OPERATEDEPT_LIST_STATISTICS',
            'columns': 'ALL',
            'source': 'WEB',
            'client': 'WEB',
            'filter': '(STATISTICSCYCLE="04")',
        }
        r = requests.get(url, params=params, timeout=10)
        data_json = r.json()
        df = pd.concat([df,pd.DataFrame(data_json['result']['data'])])
    return df



def get_operateDepartDetails(OPERATEDEPT_CODE=10063083):
    url = 'https://datacenter-web.eastmoney.com/api/data/v1/get'
    params = {
        'sortColumns': 'SECURITY_CODE,TRADE_DATE',
        'sortTypes': '-1,1',
        'pageSize': '500',
        'pageNumber': '1',
        'reportName': 'RPT_OPERATEDEPT_TRADE_DETAILSNEW',
        'columns': 'ALL',
        'source': 'WEB',
        'client': 'WEB',
        'filter': f'(OPERATEDEPT_CODE="{OPERATEDEPT_CODE}")',
        }
    pages = requests.get(url, params=params, timeout=10).json()['result']['pages']    
    
    df = pd.DataFrame()
    for i in tqdm(range(1, pages+1)):
        params = {
            'sortColumns': 'SECURITY_CODE,TRADE_DATE',
            'sortTypes': '-1,1',
            'pageSize': '500',
            'pageNumber': str(i),
            'reportName': 'RPT_OPERATEDEPT_TRADE_DETAILSNEW',
            'columns': 'ALL',
            'source': 'WEB',
            'client': 'WEB',
            'filter': f'(OPERATEDEPT_CODE="{OPERATEDEPT_CODE}")',
        }
        r = requests.get(url, params=params, timeout=10)
        data_json = r.json()
        df = pd.concat([df,pd.DataFrame(data_json['result']['data'])])
    return df
    
    
```

```python
sort_df = df[['TRADE_DATE','SECUCODE']].groupby(['SECUCODE']).max().sort_values(by=['TRADE_DATE'], ascending = False)
sort_df.columns = ['latest']
```

```python
def color_style(val):
    if val == '':
        pass
    else:
        color = 'red' if val > 0 else 'green'
        return 'color: %s' % color

df = df.fillna('')
df['TRADE_DATE'] = pd.to_datetime(df['TRADE_DATE'], format='%Y-%m-%d')
# df['ACT_BUY'] = (pd.to_numeric(df['ACT_BUY'], errors="coerce") / 100000000).round(2)
# df['ACT_SELL'] = (pd.to_numeric(df['ACT_SELL'], errors="coerce") / 100000000).round(2)
# df['NET_AMT'] = (pd.to_numeric(df['NET_AMT'], errors="coerce") / 100000000).round(2)
```
