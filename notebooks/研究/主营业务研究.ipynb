```python
spot = ak.stock_zh_a_spot_em()
```

```python
replace_words = ['的','性','业务','产品','类','收入','行业','（一）','（二）','（三）','（四）','（五）','（六）','（七）','（八）','（九）','其中'\
                ,'主营','等','系列','①','②','③','发行','出版','减：','合并抵消数','合并抵消','.','力发','�','⒓','／','“','”','分部'\
                ,'-','―','☆','%','销售','u造业','制造业','及相关','和相关','与相关','相关','制造','国内','自主','境外','品牌','大宗','贸易'\
                ,'及附','及副','及应用','服务业','及服务','与服务','和服务','服务','解决方案','板块','制品']

and_words = ['与','及','和']

#   + to ,   & to ,
```

```python
main_business = pd.DataFrame()

for i in spot['代码']:
    #print(i)
    
    try:
        stock_main_business = ak.stock_zygc_ym(symbol=i)
    except:
        print(i + ' data missing!')
    stock_mb = stock_main_business[['报告期','分类方向','分类','营业收入-占主营收入比']]
    stock_mb_prod = stock_mb[(stock_mb['分类方向'].isin(['按产品分','按行业分']) )\
             & (~stock_mb['分类'].str.contains('合计|其他|其它')) & \
            (stock_mb['营业收入-占主营收入比'].str.replace('%','').str.replace('--','0').astype(float) > 5)]
    
    for word in replace_words:
        stock_mb_prod['分类'] = stock_mb_prod['分类'].str.replace(word,'')
        
    stock_mb_prod['分类'] = stock_mb_prod['分类'].str.replace('[0-9]','').str.replace('（）','')\
                            .str.replace('、','').str.replace('：','').str.replace('()','')\
                            .str.replace(':','').str.replace(',,','').str.replace(',,','')

    
    product = ','.join(set(stock_mb_prod['分类']))

    data = {'代码' : i ,'主营' : product }
    df = pd.DataFrame(data, index=[0])
    main_business = main_business.append(df)
    
    
ls = pd.read_csv('./static/stock_industry.csv',encoding = 'gbk')
ls['代码'] = ls['股票代码'].str.replace('.SH','').str.replace('.SZ','').str.replace('.BJ','')
main_business = main_business.merge(ls, how = 'left', on = '代码')
main_business[['股票代码','股票简称','主营']].to_csv('stock_main_business.csv', index = False, encoding = 'utf-8')
```

```python
stb['主行业'] = ''
stb['副行业'] = ''

for ind in list(set(ind_list['一级行业'].tolist())):
    #print(ind)
    ind_sub_list = list(set(ind_list[ind_list['一级行业'] == ind]['一级行业'].tolist() + \
                    ind_list[ind_list['一级行业'] == ind]['二级行业'].tolist() + \
                    ind_list[ind_list['一级行业'] == ind]['三级行业'].tolist() ))
    ind_sub_list =  [x for x in ind_sub_list if x]
    #print(ind_sub_list)

    stb.loc[stb['一级行业'].isin([ind]), 'ind_sub_list'] = ','.join(ind_sub_list)

stb = stb.fillna('')
```

```python
for row, l in stb.iterrows():
    #print(l['行业-主营相关'].split(',')) 
    stb['主行业'][row] = ','.join(set(l['行业-主营相关'].split(',')).intersection((l['ind_sub_list'].split(','))))
    stb['副行业'][row] = ','.join(set(l['行业-主营相关'].split(',')).difference(l['ind_sub_list'].split(',')))
#stb.head(100)
```
