```python
index_names = {
    '上证50' :"sh000016",
    '沪深300':'sh000300',
    '中证500':'sh000905',
    '中证1000':'sh000852',
    '中证2000':'sh932000'
}

index_df = ak.stock_zh_index_daily_em(symbol="sh000001")
index_df = index_df[['date','close']].rename(columns = {'close':'上证指数'})

for i in index_names:
    df = ak.stock_zh_index_daily_em(symbol=index_names[i])
    df = df[['date','close']].rename(columns = {'close':i})
    index_df = index_df.merge(df, how='left', on = 'date')
```

```python
# plot_df = index_df[(index_df['date'] > '2005-01-01') & (index_df['date'] < '2009-07-29')] # 2008年牛市
# plot_df = index_df[(index_df['date'] > '2009-07-29') & (index_df['date'] < '2014-03-21')] # 熊市
plot_df = index_df[(index_df['date'] > '2014-03-21') & (index_df['date'] < '2016-03-01')] # 2015年牛市
```

```python
import plotly.express as px

# ['上证指数','上证50','沪深300','中证500','中证1000']

fig = px.line(plot_df, x=plot_df['date'], y=['沪深300','中证500','中证1000',])
fig.update_traces(hovertemplate="%{x|%Y/%m/%d}<br>%{y:d}<br>") 
fig.update_xaxes(
    showspikes=True,
    spikecolor="black",
    spikesnap="cursor",
    spikemode="across",
    spikedash="dot",
    spikethickness = -0.2
)
fig.show()
```

上证50和沪深300走势基本相同，可以代表A股大盘股；中证500和1000走势基本相同，可代表中盘和小盘股走势。
大盘股走势和小盘股走势从长期趋势上看大部分时间上是同涨同跌，但是在少部分时间上看，可以存在相反的走势。
