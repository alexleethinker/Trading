```python
import warnings
warnings.filterwarnings("ignore")

products_df = pd.DataFrame()


exchanges = ['EAM','EBR','EPA','XET','MIL','ELI','ATH','MAD']

for exchange in exchanges_df['hiqAbbr']:
    exchangeId = exchanges_df[exchanges_df['hiqAbbr'] == exchange]['exchangeId']
    countryCode = exchanges_df[exchanges_df['hiqAbbr'] == exchange]['countryCode']
    n_offsets = exchanges_df[exchanges_df['hiqAbbr'] == exchange]['numberOfStocks'].values[0]//1000 + 1

    for i in range(0,n_offsets):
        try:
            print(exchange, i)
            products = degiro.get_stock_list(exchangeId,countryCode, offset = i * 1000)
            products_df = products_df.append(pd.DataFrame(products))
        except:
            continue
```

```python
products_df.to_csv('degiro_allStocks.csv', index = False)
```
