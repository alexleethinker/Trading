```python
import degiroapi
from degiroapi.product import Product
from degiroapi.order import Order
from degiroapi.utils import pretty_json
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

degiro = degiroapi.DeGiro()
degiro.login("alexleethinker", "---")

exchanges = [
[663,846,'NDQ','US','NASDAQ',2922],
[676,846,'NSY','US','New York Stock Exchange',1896],
[650,846,'ASE','US','NYSE MKT - Amex',181],
[200,978,'EAM','NL','Euronext Amsterdam',127],
[320,982,'EBR','BE','Euronext Brussels',141],
[710,886,'EPA','FR','Euronext Paris',685],
[196,906,'TDG','DE','Tradegate AG',4679],
[195,906,'FRA','DE','Borse Frankfurt',1112],
[194,906,'XET','DE','Xetra',722],
]
exchanges_df = pd.DataFrame(exchanges,columns =['exchangeId','countryCode','hiqAbbr','country','name','numberOfStocks'])
```

```python
exchanges_df
```

```python
import warnings
warnings.filterwarnings("ignore")

products_df = pd.DataFrame()

for exchange in exchanges_df['hiqAbbr']:
    exchangeId = exchanges_df[exchanges_df['hiqAbbr'] == exchange]['exchangeId']
    countryCode = exchanges_df[exchanges_df['hiqAbbr'] == exchange]['countryCode']
    n_offsets = exchanges_df[exchanges_df['hiqAbbr'] == exchange]['numberOfStocks'].values[0]//1000 + 1

    for i in range(0,n_offsets):
        try:
            print(exchange, i)
            products = degiro.get_stock_list(exchangeId,countryCode, offset = i * 1000)
            products_df = products_df.append(pd.DataFrame(products))
        except:
            continue
```

```python
products_df.to_csv('degiro_allStocks.csv', index = False)
```
